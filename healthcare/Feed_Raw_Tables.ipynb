```python
from pyspark.sql import functions as F
import datetime as _dt
from pyspark.sql.types import *
```

```python
try:
    arrival_date = dbutils.widgets.get("arrival_date")
except Exception:
    arrival_date = _dt.datetime.now().strftime("%Y_%m_%d")
try:
    catalog = dbutils.widgets.get("catalog")
except Exception:
    catalog = "healthcare_cat"
try:
    schema = dbutils.widgets.get("schema")
except Exception:
    schema = "default"
```

```python
df = spark.read.format("csv").option("header","true").option("inferSchema","true").load(f"/Volumes/{catalog}/{schema}/healthcare_data/diagnosis_mapping.csv")
df.write.format("delta").mode("append").saveAsTable(f"{catalog}.{schema}.diagnosis_mapping_raw")
```

```python
patients = spark.read.format("csv").option("inferSchema","true").option("header","true").load(f"/Volumes/{catalog}/{schema}/healthcare_data/patients_daily_file_{arrival_date}.csv")
display(patients)
patients.write.format("delta").mode("append").option("mergeSchema","true").saveAsTable(f"{catalog}.{schema}.patients_daily_file_raw")
```
